```python
# importing required libraries
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# read the train and test dataset
train_data = pd.read_csv('C:/Users/Nishi/Desktop/excelR/assignments/Naive Bayes/SalaryData_Train.csv')
test_data = pd.read_csv('C:/Users/Nishi/Desktop/excelR/assignments/Naive Bayes/SalaryData_Test.csv')





```

```python
# shape of the dataset
print('Shape of training data :',train_data.shape)
print('Shape of testing data :',test_data.shape)
```

```python
train_data
```

```python
train_data.drop(['education','maritalstatus','relationship','race'], inplace = True, axis = 1)
```

```python
test_data.drop(['education','maritalstatus','relationship','race'], inplace = True, axis = 1)
```

```python
train_data
```

```python
train_data['occupation'].unique()
```

```python
data2 = pd.get_dummies(train_data, columns =['workclass', 'occupation', 'sex', 'native'])
data2
```

```python
data3 = pd.get_dummies(test_data, columns =['workclass', 'occupation', 'sex', 'native'])
data3
```

```python
# Now, we need to predict the missing target variable in the test data
# target variable - Survived

# seperate the independent and target variable on training data
train_x = data2.drop(columns=['Salary'],axis=1)
train_y = data2['Salary']
```

```python
train_y
```

```python
# seperate the independent and target variable on testing data
test_x = data3.drop(columns=['Salary'],axis=1)
test_y = data3['Salary']
```

```python
model = GaussianNB()
```

```python
# fit the model with the training data
model.fit(train_x,train_y)
```

```python
# predict the target on the train dataset
predict_train = model.predict(train_x)
print('Target on train data',predict_train) 
```

```python
# Accuray Score on train dataset
accuracy_train = accuracy_score(train_y,predict_train)
print('accuracy_score on train dataset : ', accuracy_train)

```

```python
# predict the target on the test dataset
predict_test = model.predict(test_x)
print('Target on test data',predict_test) 

# Accuracy Score on test dataset
accuracy_test = accuracy_score(test_y,predict_test)
print('accuracy_score on test dataset : ', accuracy_test)
```
